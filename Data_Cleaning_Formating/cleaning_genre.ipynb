## Cleaning the genre columns
There's too many genres and sub-genres included in the data. We will merge the sub-genre together maintain a big genre only.

```python
import pandas as pd
df = pd.read_csv("Database_with_lyrics.csv")
df =df.drop(columns=["Unnamed: 0"])
df[::2]
```

```python
print("The number of genres originally: ", len(df["genre"].unique()))
print("\nThe genres originally: ",df["genre"].unique() )
```

```python
genres_check = ['pop', 'rock', 'hip hop', 'rap', 'jazz', 'house', 'blues', 'r&b', 'folk', 'metal', 'drill', 'funk', 'techno', 'country', 'country', 'ballad', 'phonk', 'punk', 'club', 'reggae', 'dance', 'disco', 'emo', 'classical', 'indie', 'grunge', 'trap']
def generalize_genre(specific_genre):
    if not isinstance(specific_genre, str):
        return specific_genre
    for genre in genres_check:
        if genre in specific_genre:
            return genre
    return specific_genre
    
df['general_genre'] = df['genre'].apply(generalize_genre)
df = df.drop(columns="genre")
df["genre"] = df["general_genre"]
df = df.drop(columns="general_genre")
print(len(df["genre"].unique()))
df["genre"].unique()
```
