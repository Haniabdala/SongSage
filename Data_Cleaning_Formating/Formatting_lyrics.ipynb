```python
import pandas as pd
```

```python
df = pd.read_csv("Database.csv")
```

### Lyrics pre-processing

```python
lyrics = df['Lyrics']

for i in range(len(lyrics)):
    lyrics[i] = str(lyrics[i]) # Casting to String to avoid errors
    lyrics[i] = lyrics[i].lower() #To LowerCase to ensure that the word counting is not redundant for the same word
    #Removing Special Characters to ensure error-free lyrics processing
    lyrics[i] = lyrics[i].replace("\'", "")
    lyrics[i] = lyrics[i].replace("\"", "")
    lyrics[i] = lyrics[i].replace("\xa0","")

lyrics
```

```python
df['Lyrics'] = lyrics
df.to_csv("Database_cleaned_lyrics.csv")
```

### Genre Merging (artist and song genre)

```python
song_genre = df['genre']
artist_genre = df['artist_genre']
```

```python
for i in range(len(artist_genre)):
    if(artist_genre[i] == 'Not Found'):
        artist_genre[i] = song_genre[i]
```

```python
df_cleaned_genre = df.drop("genre",axis='columns')
df_cleaned_genre.to_csv("Database_cleaned_genre.csv")
```
